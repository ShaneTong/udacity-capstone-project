```python
path = '../../us-financial-news-articles'
os.listdir(path)
corpus = []

for subdir in os.listdir(path):
    if subdir != '.DS_Store':
        cur_path = path+'/'+subdir
        for filename in os.listdir(cur_path):
            if filename.endswith('.json'):
                file_path = cur_path+'/'+filename
                file_dict = json.load(open(file_path))
                corpus.append({"text": file_dict['text'], "published": file_dict['published']})
```

```python
company_ticker_map = {'google': 'GOOG',
                     'amazon': 'AMZN',
                     'facebook': 'FB',
                     'microsoft': 'MSFT'}
for company, art_list in corpus_dict.items():
    price_mat = pdr.data.YahooDailyReader(company_ticker_map[company], '2018-01-01').read()
    for art_info in art_list:
        news_date = art_info['published'].split('T')[0]
        art_info['date'] = news_date
        try:
            start_price = price_mat['Open'][news_date]
            end_price = price_mat['Close'][news_date]
            art_info['percentage_change'] = (end_price - start_price)/start_price
        except:
            art_info['percentage_change'] = None
```
